```python
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoding
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['ProductName']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['ProductName']))
final_df = df.join(encoded_df)
final_df.drop('ProductName', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Platform']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Platform']))
final_df = final_df.join(encoded_df)
final_df.drop('Platform', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Processor']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Processor']))
final_df = final_df.join(encoded_df)
final_df.drop('Processor', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['SkuEdition']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['SkuEdition']))
final_df = final_df.join(encoded_df)
final_df.drop('SkuEdition', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_MDC2FormFactor']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_MDC2FormFactor']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_MDC2FormFactor', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_PrimaryDiskTypeName']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_PrimaryDiskTypeName']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_PrimaryDiskTypeName', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_ChassisTypeName']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_ChassisTypeName']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_ChassisTypeName', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_PowerPlatformRoleName']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_PowerPlatformRoleName']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_PowerPlatformRoleName', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_OSBranch']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_OSBranch']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_OSBranch', axis=1, inplace=True)

encoded = encoder.fit_transform(final_df[['Census_OSEdition']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_OSEdition']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_OSEdition', axis=1, inplace=True)



encoded = encoder.fit_transform(final_df[['Census_GenuineStateName']])
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['Census_PrimaryDiskTypeName']))
final_df = final_df.join(encoded_df)
final_df.drop('Census_PrimaryDiskTypeName', axis=1, inplace=True)
```

```python
final_df.to_csv("test.csv")
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop('ImmersionLevel',axis=1)
y = final_df['ImmersionLevel']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score , ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)

print("--Decision Trees--\n")
print(" confusion matrix : \n",confusion_matrix(y_test,predictions) , "\n")
print("accuracy : ",accuracy_score(y_test,predictions)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 20, criterion = 'gini', max_depth = 4) 
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)

print("--Random Forests--\n")
print(" confusion matrix : \n",confusion_matrix(y_test,rfc_pred) , "\n")
print("accuracy : ",accuracy_score(y_test,rfc_pred)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
 
y_pred = gnb.predict(X_test)

print("--Naive Bayes Classifiers--\n")
print(" confusion matrix : \n",confusion_matrix(y_test,y_pred) , "\n")
print("accuracy : ",accuracy_score(y_test,y_pred)*100)
```
